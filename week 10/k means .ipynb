```python
import pandas as pd
from sklearn.cluster import KMeans

# Create sample dataset (no CSV needed)
data = pd.DataFrame({
    'x': [1, 2, 1, 6, 7, 8],
    'y': [1, 1, 2, 6, 7, 8]
})

# Features
X = data[['x', 'y']]

# Apply KMeans
kmeans = KMeans(n_clusters=2, random_state=42, n_init=10)
data['Cluster'] = kmeans.fit_predict(X)

# Output
print("Clustered Data:\n", data)
print("\nCentroids:\n", kmeans.cluster_centers_)
```
